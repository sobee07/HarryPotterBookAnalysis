```python
# Import libraries
import glob
import pandas as pd
```

```python
Use glob to get a list of filenames
```

```python
import os
path = 'data'
files= glob.glob(path + '/*.txt')
list = []
for f in files:
    # get filename
    stock = os.path.basename(f)
    list.append(stock)
print (list)    
```

```python
list_contents=[]
for f in list:
    # get filename
    myfile = open(f, 'r',encoding='utf-8')   
    reading_file=myfile.read()    
    list_contents.append(reading_file)
    myfile.close()

```

```python
print(list_contents)
```

```python
Create a DataFrame containing the 
```

```python
df= pd.DataFrame()
df['name'] = list
df['full_text'] = list_contents
df
```
